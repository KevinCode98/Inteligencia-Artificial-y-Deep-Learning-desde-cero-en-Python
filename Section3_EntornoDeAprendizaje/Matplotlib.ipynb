## Representación gráfica de datos

si la función de trazado se le da un matriz de datos, la usará como coordenadas en el eje vertical, y utilizará el índice de cada punto de datos en el array como la coordenada horizontal.

```python
lista = [1,2,5,7,8,3,1]

plt.plot(lista)
plt.show()
```

También se puede proporcionar dos matrices: una para el eje horizontal, y otra para el eje vertical.

```python
lista2 = [1,4,6,7,8,9,10]
plt.plot(lista2, lista)
plt.show()
```

Pueden modificarse la longitudes de los ejes para que la figura no se vea tan ajustada.

```python
plt.plot([-3,-1,0,4,7], [1,4,6,7,8])
plt.show()
```

```python
plt.plot([-3,-1,0,4,7], [1,4,6,7,8])
plt.axis([-4,8,0,10])
plt.show()
```

Se sigue el mismo procedimiento para pintar una función matemática.

```python
import numpy as np

x = np.linspace(-2, 2, 500)
y = x ** 2

plt.plot(x,y)
plt.show
```

También puede modificarse el estilo de la gráfica par aque contenga más información.

```python
plt.plot(x,y)
plt.title("Squere function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

Pueden superponerse gráficas y cambiar el estilo de las líneas

```python
import numpy as np

x = np.linspace(-2, 2, 500)
y = x ** 2
y2 = x + 1

plt.plot(x,y,'b--', x,y2, 'g')
plt.show()
```

```python
import numpy as np

x = np.linspace(-2, 2, 500)
y = x ** 2
y2 = x + 1

plt.plot(x,y, 'b--')
plt.plot(x,y2, 'g')
plt.grid(True)
plt.show()
```

Para poder diferencia entra ambas funciones siempre es recomendable añadir una leyenda

```python
import numpy as np

x = np.linspace(-2, 2, 500)
y = x ** 2
y2 = x + 1

plt.plot(x,y, 'b--', label="x**2")
plt.plot(x,y2, 'g', label="x+1")
plt.legend(loc="best")
plt.show()
```

Tambien puede crearse dos gráficas que no se superpongan. Estas graficas se organizan en un griid y se denominan subplots.

```python
x = np.linspace(-2, 2, 500)
y = x ** 2
y2 = x + 1

plt.subplot(1, 2, 1)
plt.plot(x,y, 'b--')

plt.subplot(1, 2, 2)
plt.plot(x,y2, 'g')
plt.show()
```

Para que las gráficas no quden tan ajustadas, podemos hacer la figura más grande.

```python
plt.figure(figsize=(14,6))

x = np.linspace(-2, 2, 500)
y = x ** 2
y2 = x + 1

plt.subplot(1, 2, 1)
plt.plot(x,y, 'b--')
plt.xlabel("x1", fontsize=14)
plt.ylabel("y1", fontsize=14)

plt.subplot(1, 2, 2)
plt.plot(x,y2, 'g')
plt.xlabel("x2", fontsize=14)
plt.ylabel("y2", fontsize=14)
plt.show()
```

## Scatter plots

```python
from numpy.random import rand

x,y = rand(2, 100)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from numpy.random import rand

x,y = rand(2, 100)
x2,y2 = rand(2, 100)

plt.scatter(x,y,c="red")
plt.scatter(x2,y2,c="blue")
```

## Histograma

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]

plt.subplot(211)
plt.xlabel("Value")
plt.ylabel("Frecuency")
plt.hist(data, bins=15, rwidth=0.8)
# ----------------------------------------------

plt.subplot(212)
plt.hist(data, bins=5, rwidth=0.8)
plt.xlabel("Value")
plt.ylabel("Frecuency")
plt.show()
```
